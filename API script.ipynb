## Handling the Artists column
To get the information present in track.artists, we need to flatten the dataframe, creating duplicated rows for when multiple artists collaborate on a song. This will enable us to create a table with only the ids later in SQL. 
To do so, we use the explode function, and then we apply the json_normalize to retrieve the artists information with extra columns.

```python

exploded_df = track_info.explode('track.artists').reset_index(drop=True)
normalized_df = pd.json_normalize(exploded_df['track.artists'])
master_df = pd.concat([exploded_df.drop(columns=['track.artists']), normalized_df], axis=1)
master_df.head()
```

```python
# From the created columns, we don't want to keep all of them 
master_df.drop(columns=['href','type','uri'], inplace=True)
```

### Available Markets
There are two columns with lists of available_markets. One corresponds to where the album is available, and the other one where the track is. In getting only the intersection of those two lists, we will create an element for the table 'album_artist_track' with only the markets where the song and the associated album are available. 

```python
def intersection(row):
    return list(set(row["track.album.available_markets"]) & set(row["track.available_markets"]))
```

```python
master_df['available_markets'] = master_df.apply(intersection, axis=1)
master_df.head()
```
